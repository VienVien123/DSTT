l·∫ßn l∆∞·ª£t 2, 3, 4 gi·ªù sau l∆∞·ª£ng thu·ªëc c√≤n l·∫°i l√† 7, 5 v√† 2.
H√£y vi·∫øt c√°c l·ªánh b·∫±ng Python ƒë·ªÉ t√¨m ph∆∞∆°ng tr√¨nh tuy·∫øn t√≠nh b·∫±ng ph∆∞∆°ng ph√°p b√¨nh ph∆∞∆°ng
c·ª±c ti·ªÉu v·ªõi t·∫≠p c√°c c·∫∑p s·ªë (ùë•, ùë¶) ƒë·ªÉ t√¨m c√¥ng th·ª©c gi·∫£m l∆∞·ª£ng thu·ªëc ƒë·ªëi v·ªõi b·ªánh nh√¢n theo th·ªùi
gian:

 $(ùë•, ùë¶) = {(0, 10), (1, 8), (2, 7), (3, 5), (4, 2)}$

[K·∫øt qu·∫£ tham kh·∫£o ƒë·ªÉ ki·ªÉm ch·ª©ng: Ph∆∞∆°ng tr√¨nh tuy·∫øn t√≠nh l√†: ùë¶ =[ 10.2 ‚àí 1.9ùë•]

```python
x = np.array([[1,0],[1,1],[1,2],[1,3],[1,4]])
y = np.array([[10],[8],[7],[5],[2]])
A = np.linalg.inv((x.T)@x)@((x.T)@y)
print(A)
```

```python
x_data = [0, 1, 2, 3, 4]
y_data = [10, 8, 7, 5, 2]
a = np.vstack([x_data, np.ones(len(x_data))]).T
b = np.array(y_data)

A = np.linalg.lstsq(a, b, rcond=None)[0]
fig, ax = plt.subplots()
ax.axhline(y=0, color='black', alpha=0.5)
ax.axvline(x=0, color='black', alpha=0.5)
x_values = np.linspace(min(x_data) - 1, max(x_data) + 1, 100)
y_values = A[0] * x_values + A[1]
ax.plot(x_values, y_values, color='g', linewidth=3, label='Fitted line', alpha=0.85)
ax.plot(x_data, y_data, 'o', c='red', lw=3, alpha=0.85, label='Data points')
ax.set_title('Fitted Line')
ax.legend(loc='upper left')
plt.show()
```

# **B√ÄI 2**
Cho m√¥ h√¨nh $ y = a_0 + a_1x + a_2x^2 $. T√¨m c√°c tham s·ªë $a_0, a_1$ v√† $a_2$ ƒë·ªÉ m√¥ h√¨nh ph√π h·ª£p v·ªõi t·ª´ng b·ªô d·ªØ li·ªáu sau.

a/

$\begin{array}{|c|c|c|c|c|}
\hline
x&2&3&5&6\\
\hline
y&0&-10&-48&-76\\
\hline
\end{array}$

ta c√≥:

(x,y)= (2,0),(3,-10),(5,-48),(6,-76)

y= (0,-10,-48,-76)

```python
x=np.array([[1,2,4],[1,3,9],[1,5,25],[1,6,36]])
y=np.array([0,-10,-48,-76])
A= linalg.lstsq(x,y)[0]
print(A)
```

```python
a = np.array([[2, 3, 5, 6]])
b = np.array([0, -10, -48, -76])
fig, ax = plt.subplots()
ax.axhline(y=0, color='black', alpha=0.5)
ax.axvline(x=0, color='black', alpha=0.5)
c = np.linspace(-10, 11, 100)
bb = A[0] + A[1] * c + A[2] * c *c
ax.plot(c, bb, color='g', linewidth=3, label='Fitted Parabola', alpha=0.85)
ax.plot(a[0], b, 'o', c='red', lw=3, alpha=0.85, label='Data points')
ax.set_xlim(a.min() -5, a.max() + 5)
ax.set_ylim(b.min() -5, b.max() + 5)
ax.set_title('Fitted Parabola')
ax.legend(loc='upper left')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.axhline(y=0, color='black', alpha=0.5)
ax.axvline(x=0, color='black', alpha=0.5)

a = np.linspace(-10, 11, 100)
bb = B[0] + B[1] * a + B[2] * a ** 2  # y = ax^2 + bx + c
ax.plot(a, bb, color='g', linewidth=3, label='Fitted Parabola', alpha=0.85)
ax.plot(x[:,1], y, 'o', c='red', lw=3, alpha=0.85, label='Data points')
ax.set_xlim(x.min() - 1, x.max() + 1)
ax.set_ylim(y.min() - 1, y.max() + 1)
ax.set_title('Fitted Parabola')
ax.legend(loc='upper left')

plt.show()
```

# **B√ÄI 3**

Cho m√¥ h√¨nh $y = a + \frac{b}{x}$. t√¨m c√°c tham s√¥ a v√† b ƒë·ªÉ m√¥ h√¨nh ph√π h·ª£p v·ªõi b·ªô d·ªØ li·ªáu sau:

```python
x=np.array([[1,1],[1,1/3],[1,1/6]])
y= np.array([7,3,1])
c= linalg.lstsq(x,y)[0]
print(c)
```

```python
x = np.array([1, 3, 6])
y = np.array([7, 3, 1])
fig, ax = plt.subplots()
ax.axhline(y=0, color='black', alpha=0.5)
ax.axvline(x=0, color='black', alpha=0.5)
a = np.linspace(0.1, 11, 100)
bb = c[0] + c[1] / a # y = ax + b
ax.plot(a, bb, color='g', linewidth=3, label='Fitted Parabola', alpha=0.85)
ax.plot(x, y, 'o', c='red', lw=3, alpha=0.85, label='Data points')
ax.set_xlim(x.min() - 1, x.max() + 1)
ax.set_ylim(y.min() - 1, y.max() + 1)
ax.set_title('Fitted Parabola')
ax.legend()
plt.show()
```

# **B√ÄI 4**

cho m√¥ h√¨nh $y = a + b \sqrt{x}$. T√¨m c√°c tham s·ªë a v√† b ƒë·ªÉ m√¥ h√¨nh ph√π h·ª£p v·ªõi b·ªô d≈© li·ªáu sau:

$\begin{array}{|c|c|c|c|}
\hline
x&3&7&10\\
\hline
y&1.5&2.5&3\\
\hline
\end{array}$


```python
X= np.array([[1,np.sqrt(3)],[1,np.sqrt(7)],[1,np.sqrt(10)]])
Y = np.array([1.5,2.5,3])
C=linalg.lstsq(X,Y)[0]
print(C)
```

```python
x = [3, 7, 10]
y = [0.5, 2.5, 3]

fig, ax = plt.subplots()
ax.axhline(y=0, color='black', alpha=0.5)
ax.axvline(x=0, color='black', alpha=0.5)
a = np.linspace(0.1, 11, 100)
bb = C[0] + C[1] * np.sqrt(a)  # y = a + b*sqrt(x)
ax.plot(a, bb, color='g', linewidth=3, label='Fitted Parabola', alpha=0.85)
ax.plot(x, y, 'o', c='red', lw=3, alpha=0.85, label='Data dot')
```
