## Bai 2

Cho mo hinh $\text{y} = \alpha_0 + \alpha_1\text{x} + \alpha_2\text{x}^2$. Tim cac tham so $\alpha_0, \alpha_1 , \alpha_2$ de mo hinh phu hop voi tung bo du lieu sau:

$\text{a)}\\$
$\begin{array}{|c|c|c|c|c|}
\hline
x&2&3&5&6\\
\hline
y&0&-10&-48&-76\\
\hline
\end{array}$

```python
#a
X = np.array([[1,2,4],[1,3,9],[1,5,25],[1,6,36]])
Y = np.array([[0],[-10],[-48],[-76]])
A = (np.linalg.inv((X.T)@X))@((X.T)@Y)
a0 = float(A[0][0])
a1 = float(A[1][0])
a2 = float(A[2][0])
print(f"a0 = {a0}")
print(f"a1 = {a1}")
print(f"a2 = {a2}")
```

```python
#Ve do thi
plt.figure(figsize=(10, 5))
X = np.linspace(-10, 10, 400)
Y = a0 + a1*X + a2*X**2
plt.xlabel("X") 
plt.ylabel("Y")  
plt.plot(X, Y, label='Mo hinh y = {:.2f} + {:.2f}x + {:.2f}x^2'.format(a0, a1, a2))
plt.legend() 
plt.show()
```

$\text{b)}\\$
$\begin{array}{|c|c|c|c|c|}
\hline
x&1&0&1&2\\
\hline
y&-2&-1&0&4\\
\hline
\end{array}$

```python
#b
X = np.array([[1,1,1],[1,0,0],[1,1,1],[1,2,4]])
Y = np.array([[-2],[-1],[0],[4]])
A = (np.linalg.inv((X.T)@X))@((X.T)@Y)
a0 = float(A[0][0])
a1 = float(A[1][0])
a2 = float(A[2][0])
print(f"a0 = {a0}")
print(f"a1 = {a1}")
print(f"a2 = {a2}")
```

```python
plt.figure(figsize=(10,5))
X = np.linspace(-10,10, 400)
Y = a0 + a1*X + a2*X**2
plt.xlabel("X") 
plt.ylabel("Y")  
plt.plot(X, Y, label='Mo hinh y = {:.2f} + {:.2f}x + {:.2f}x^2'.format(a0, a1, a2))
plt.legend() 
plt.show()
```

## Bai 3


$\begin{array}{|c|c|c|c|}
\hline
x&1&3&6\\
\hline
y&7&3&1\\
\hline
\end{array}\\$
$\text{y} = a + \frac{b}{x}$

```python
#c1
X = np.array([[1,1],[1,1/3],[1,1/6]])
Y = np.array([[7],[3],[1]])
A = np.linalg.inv(np.transpose(X)@X)@(np.transpose(X)@Y)
a = A[0][0]
b = A[1][0]
print(f"a = {a}")
print(f"b = {b}")
```

```python
plt.figure(figsize=(10,5))
X = np.linspace(-10,10,400)
Y = a + b/X
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(X, Y, label='Mo hinh y = {:.2f} + {:.2f}/x'.format(a,b))
plt.legend()
plt.show()
```

## Bai 4



$\begin{array}{|c|c|c|c|}
\hline
x&3&7&10\\
\hline
y&1.5&2.5&3\\
\hline
\end{array}\\$
$\text{y} = a + b\sqrt{x}$

```python
X = np.array([[1,np.sqrt(3)],[1,np.sqrt(7)],[1,np.sqrt(10)]])
Y = np.array([[1.5],[2.5],[3]])
A = np.linalg.inv(np.transpose(X)@X)@(np.transpose(X)@Y)
a = A[0][0]
b = A[1][0]
print(f"a = {a}")
print(f"b = {b}")
```

```python
plt.figure(figsize = (10,5))
a = float(A[0][0])
a1 = float(A[1][0])
X = np.linspace(0, 100, 400)
Y = a + a1 * np.sqrt(X)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(X,Y,label='Mo hinh y = {:.2f} + {:.2f}x'.format(a,b))
plt.legend()
plt.show()
```
